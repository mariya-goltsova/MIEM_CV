```python
cap = cv.VideoCapture(video_path)
video_len = int(cap.get(cv.CAP_PROP_FRAME_COUNT))
```

```python
person1 = f'drive/MyDrive/immm/miem_faces/1_miem.jpeg'
person2 = f'drive/MyDrive/immm/miem_faces/491_miem.jpeg'

tp = 0
fp = 0
flag = 0
step = 50
```

```python
while flag < video_len:
  print(flag, tp, fp)

  cap.set(cv.CAP_PROP_POS_FRAMES, flag)
  ret, frame = cap.read()

  try:
    extract_faces = DeepFace.extract_faces(frame)
  except ValueError:
    flag += step
    continue


  for face in extract_faces:
    x = face["facial_area"]["x"]
    y = face["facial_area"]["y"]

    dx = face["facial_area"]["w"]
    dy = face["facial_area"]["h"]
    try:
      is_ok = False

      if 7600 <= flag <= 7750: # only for 1 person
        is_ok = is_ok or DeepFace.verify(person1, frame[y:y+dy, x:x+dx])["verified"]
      else:
        is_ok = is_ok or DeepFace.verify(person1, frame[y:y+dy, x:x+dx])["verified"]
        is_ok = is_ok or DeepFace.verify(person2, frame[y:y+dy, x:x+dx])["verified"]
      if is_ok:
        tp += 1
      else:
        fp += 1

    except ValueError:
      fp += 1
  flag += step
cap.release()
```

```python
def print_metric(TP, FP):
  precision = (float(TP) / (float(TP) + float(FP)))
  return np.round(precision, 3)
```

```python
print_metric(tp, fp)
```
